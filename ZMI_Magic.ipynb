````python
import math

# --- Inputs -------------------------------------------------------------

OptimalPath_XP = {
    1: 7.2032, 2: 7.2910, 3: 7.3638, 4: 7.4252,
    5: 10.7646, 6: 10.8396, 7: 10.9056, 8: 10.9642,
    9: 14.2942, 10: 14.3574, 11: 14.4132, 12: 14.4654,
    13: 17.2566, 14: 17.3034, 15: 17.3516, 16: 17.3902,
    17: 17.4302, 18: 17.4646
}

CheapPath_XP = {
    1: 7.2032, 2: 7.2910, 3: 7.3638, 4: 7.4252,
    5: 8.2947, 6: 8.3300, 7: 8.3617, 8: 8.3907,
    9: 8.7971, 10: 8.8298, 11: 8.8583, 12: 8.8847,
    13: 9.2566, 14: 9.2830, 15: 9.3090, 16: 9.3327,
    17: 9.3565, 18: 9.3783
}

# XP to go from level N -> N+1 (we'll only use 1–15 here)
XPToNext = {
    1: 83, 2: 91, 3: 102, 4: 112, 5: 124,
    6: 138, 7: 151, 8: 168, 9: 185, 10: 204,
    11: 226, 12: 249, 13: 274, 14: 304,
    15: 335, 16: 369, 17: 408
}

LEVELS = range(1, 16)  # up to level 15


# --- Effective level from prayers --------------------------------------

def lore_effective_level(lvl: int) -> int:
    """Mystic Lore: effective level floor(1.1 * lvl)."""
    return math.floor(1.1 * lvl)

def might_effective_level(lvl: int) -> int:
    """Mystic Might: effective level floor(1.15 * lvl)."""
    return math.floor(1.15 * lvl)


# --- XP-per-cast lookup based on path + prayer mode --------------------

def xp_per_cast(path: str, lvl: int, mode: str) -> float:
    """
    path: 'opt' or 'cheap'
    mode: 'none', 'lore', or 'might'

    We use the 'effective level' to index into the XP tables.
    """
    if path == "opt":
        table = OptimalPath_XP
    else:
        table = CheapPath_XP

    if mode == "none":
        eff_lvl = lvl
    elif mode == "lore":
        eff_lvl = lore_effective_level(lvl)
    elif mode == "might":
        eff_lvl = might_effective_level(lvl)
    else:
        raise ValueError("Unknown mode")

    return table[eff_lvl]


# --- Compute casts per level & cumulative for each combo ----------------

def compute_casts(mode: str, path: str):
    """Return (casts_per_level, cumulative_casts) dicts for a given prayer mode and path."""
    casts_per_level = {}
    cumulative = {}

    cum_raw = 0.0  # running expected total (unceded)

    for lvl in LEVELS:
        xp_needed = XPToNext[lvl]
        xp_cast = xp_per_cast(path, lvl, mode)

        casts_raw = xp_needed / xp_cast
        casts = math.ceil(casts_raw)  # integer casts to next level

        casts_per_level[lvl] = casts

        cum_raw += casts_raw
        cumulative[lvl] = math.ceil(cum_raw)  # ceil the running total

    return casts_per_level, cumulative


# Compute all 6 forks
NoPrayer_OptPerLevel,   NoPrayer_OptCum   = compute_casts("none",  "opt")
NoPrayer_CheapPerLevel, NoPrayer_CheapCum = compute_casts("none",  "cheap")

Lore_OptPerLevel,       Lore_OptCum       = compute_casts("lore",  "opt")
Lore_CheapPerLevel,     Lore_CheapCum     = compute_casts("lore",  "cheap")

Might_OptPerLevel,      Might_OptCum      = compute_casts("might", "opt")
Might_CheapPerLevel,    Might_CheapCum    = compute_casts("might", "cheap")


# --- Example pretty print (optional) -----------------------------------

print("Lvl | XP->Next | NoOpt | NoOptCum | NoCheap | NoCheapCum | LoreOpt | LoreOptCum | LoreCheap | LoreCheapCum | MightOpt | MightOptCum | MightCheap | MightCheapCum")
print("-" * 150)
for lvl in LEVELS:
    print(
        f"{lvl:3d} | "
        f"{XPToNext[lvl]:8d} | "
        f"{NoPrayer_OptPerLevel[lvl]:5d} | "
        f"{NoPrayer_OptCum[lvl]:8d} | "
        f"{NoPrayer_CheapPerLevel[lvl]:7d} | "
        f"{NoPrayer_CheapCum[lvl]:10d} | "
        f"{Lore_OptPerLevel[lvl]:7d} | "
        f"{Lore_OptCum[lvl]:11d} | "
        f"{Lore_CheapPerLevel[lvl]:9d} | "
        f"{Lore_CheapCum[lvl]:12d} | "
        f"{Might_OptPerLevel[lvl]:8d} | "
        f"{Might_OptCum[lvl]:12d} | "
        f"{Might_CheapPerLevel[lvl]:11d} | "
        f"{Might_CheapCum[lvl]:13d}"
    )

print("\nNoPrayer_OptPerLevel =", NoPrayer_OptPerLevel)
print("NoPrayer_CheapPerLevel =", NoPrayer_CheapPerLevel)
print("Lore_OptPerLevel =", Lore_OptPerLevel)
print("Lore_CheapPerLevel =", Lore_CheapPerLevel)
print("Might_OptPerLevel =", Might_OptPerLevel)
print("Might_CheapPerLevel =", Might_CheapPerLevel)

print("\nNoPrayer_OptCum =", NoPrayer_OptCum)
print("NoPrayer_CheapCum =", NoPrayer_CheapCum)
print("Lore_OptCum =", Lore_OptCum)
print("Lore_CheapCum =", Lore_CheapCum)
print("Might_OptCum =", Might_OptCum)
print("Might_CheapCum =", Might_CheapCum)
````

````python
import math

rune_yields = {
    "1-9":    {"Fire": 3.00, "Earth": 6.00, "Water": 12.00, "Mind": 25.00, "Air": 50.00, "XP_Ess": 9.39},
    "10-19":  {"Fire": 12.00, "Earth": 24.00, "Water": 21.00, "Mind": 18.00, "Air": 15.00, "XP_Ess": 10.52},
    "20-29":  {"Fire": 15.00, "Earth": 14.00, "Water": 13.50, "Mind": 13.00, "Air": 12.00, "XP_Ess": 11.34},
    "30-39":  {"Fire": 12.00, "Earth": 11.00, "Water": 9.00,  "Mind": 8.00,  "Air": 7.00,  "XP_Ess": 12.27},
    "40-49":  {"Fire": 8.00,  "Earth": 7.50, "Water": 7.00,  "Mind": 6.50,  "Air": 6.00,  "XP_Ess": 12.90},
    "50-59":  {"Fire": 7.00,  "Earth": 6.50, "Water": 6.00,  "Mind": 5.50,  "Air": 5.00,  "XP_Ess": 13.47},
}

xp_for_level = {
    1: 0,    2: 83,    3: 174,   4: 276,   5: 388,
    6: 512,  7: 650,   8: 801,   9: 969,   10: 1154,
    11: 1358, 12: 1584, 13: 1833, 14: 2107, 15: 2411,
    16: 2746, 17: 3115, 18: 3520, 19: 3973, 20: 4470,
    21: 5018, 22: 5624, 23: 6291, 24: 7028, 25: 7842,
    26: 8740, 27: 9730, 28: 10824, 29: 12031, 30: 13363,
    31: 14833, 32: 16456, 33: 18247, 34: 20224, 35: 22406,
    36: 24815, 37: 27473, 38: 30403, 39: 33648, 40: 37224,
    41: 41171, 42: 45529, 43: 50343, 44: 55649, 45: 61512,
    46: 67983, 47: 75127, 48: 83021, 49: 91721, 50: 101333,
}

MAX_LEVEL = max(xp_for_level.keys())


def get_rune_block(level: int) -> dict:
    """Return the rune_yields entry for the given level."""
    if 1 <= level <= 9:
        return rune_yields["1-9"]
    elif 10 <= level <= 19:
        return rune_yields["10-19"]
    elif 20 <= level <= 29:
        return rune_yields["20-29"]
    elif 30 <= level <= 39:
        return rune_yields["30-39"]
    elif 40 <= level <= 49:
        return rune_yields["40-49"]
    else:  # 50–59 (and above, capped)
        return rune_yields["50-59"]

def simulate_crafting(total_essence: float,
                      batch_size: int = 25,
                      start_level: int = 1,
                      start_xp: float | None = None):
    """
    Simulate crafting with up to `batch_size` essence at a time.
    Returns (final_level, final_xp (floored), rune_counts_dict (floored)).
    """
    # Initial state
    level = start_level
    xp = xp_for_level[level] if start_xp is None else start_xp
    ess_left = float(total_essence)

    # Float values internally (floored at the end)
    runes = {"Fire": 0.0, "Earth": 0.0, "Water": 0.0, "Mind": 0.0, "Air": 0.0}

    while ess_left > 1e-9 and level < MAX_LEVEL:
        ess_this_trip = min(batch_size, ess_left)

        block = get_rune_block(level)
        xp_per_ess = block["XP_Ess"]

        if level == MAX_LEVEL:
            ess_used = ess_this_trip
        else:
            xp_to_next = xp_for_level[level + 1] - xp
            ess_to_next = xp_to_next / xp_per_ess if xp_per_ess > 0 else float("inf")
            ess_used = min(ess_this_trip, ess_to_next)

        # Gain runes for ess_used at this level
        for r in ["Fire", "Earth", "Water", "Mind", "Air"]:
            runes[r] += ess_used * (block[r] / 100.0)

        # Gain xp
        xp += ess_used * xp_per_ess

        ess_left -= ess_used

        # Level up if XP crosses thresholds
        while level < MAX_LEVEL and xp >= xp_for_level[level + 1]:
            level += 1

    # ---- Apply flooring here ----
    xp = math.floor(xp)

    for r in runes:
        runes[r] = math.floor(runes[r])

    return level, xp, runes
````

````python
import math
import random
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
# DATA TABLES
# ----------------------------------------------------------------------

# XP Per Strike Cast: {damage_done: xp_gained}
WindStrike_XP  = {0: 5.5, 1: 7.5, 2: 9.5, 3: 11.5, 4: 13.5, 5: 15.5}
WaterStrike_XP = {0: 7.5, 1: 9.5, 2: 11.5, 3: 13.5, 4: 15.5, 5: 17.5}
# Earth now extended to 10 max hit
EarthStrike_XP = {
    0: 9.5, 1: 11.5, 2: 13.5, 3: 15.5, 4: 17.5,
    5: 19.5, 6: 21.5, 7: 23.5, 8: 25.5, 9: 27.5, 10: 29.5
}
FireStrike_XP  = {0: 11.5, 1: 13.5, 2: 15.5, 3: 17.5, 4: 19.5, 5: 21.5}

SPELL_XP_TABLE = {
    "Wind":  WindStrike_XP,
    "Water": WaterStrike_XP,
    "Earth": EarthStrike_XP,
    "Fire":  FireStrike_XP,
    # Weaken is handled separately (flat XP)
}

WEAKEN_XP = 21.0  # flat xp per cast

# Hit chance: {effective_magic_level: {damage: probability}}
# This remains your ORIGINAL generic table (applies to Wind/Water/Fire).
Lvl = {
    1: {0: 0.3613, 1: 0.4258, 2: 0.2129},
    2: {0: 0.3284, 1: 0.4477, 2: 0.2239},
    3: {0: 0.3011, 1: 0.4659, 2: 0.2330},
    4: {0: 0.2780, 1: 0.4814, 2: 0.2406},

    5: {0: 0.2581, 1: 0.2967, 2: 0.1484, 3: 0.1484, 4: 0.1484},
    6: {0: 0.2410, 1: 0.3036, 2: 0.1518, 3: 0.1518, 4: 0.1518},
    7: {0: 0.2260, 1: 0.3096, 2: 0.1548, 3: 0.1548, 4: 0.1548},
    8: {0: 0.2126, 1: 0.3150, 2: 0.1575, 3: 0.1575, 4: 0.1574},

    9:  {0: 0.2009, 1: 0.2283, 2: 0.1142, 3: 0.1142, 4: 0.1142, 5: 0.2282},
    10: {0: 0.1903, 1: 0.2314, 2: 0.1157, 3: 0.1157, 4: 0.1157, 5: 0.2312},
    11: {0: 0.1810, 1: 0.2341, 2: 0.1170, 3: 0.1170, 4: 0.1170, 5: 0.2339},
    12: {0: 0.1723, 1: 0.2365, 2: 0.1183, 3: 0.1183, 4: 0.1183, 5: 0.2363},

    13: {0: 0.1643, 1: 0.1857, 2: 0.0929, 3: 0.0929, 4: 0.0929, 5: 0.3713},
    14: {0: 0.1575, 1: 0.1873, 2: 0.0936, 3: 0.0936, 4: 0.0936, 5: 0.3744},
    15: {0: 0.1506, 1: 0.1887, 2: 0.0944, 3: 0.0944, 4: 0.0944, 5: 0.3775},
    16: {0: 0.1449, 1: 0.1901, 2: 0.0950, 3: 0.0950, 4: 0.0950, 5: 0.3800},
    17: {0: 0.1391, 1: 0.1913, 2: 0.0957, 3: 0.0957, 4: 0.0957, 5: 0.3825},
    18: {0: 0.1341, 1: 0.1925, 2: 0.0962, 3: 0.0962, 4: 0.0962, 5: 0.3848},
}

# EARTH-ONLY hit distributions for eff. levels 9–16
EarthLvl = {
    9:  {0: 0.2097, 1: 0.1976, 2: 0.0988, 3: 0.0988, 4: 0.0988,
         5: 0.0988, 6: 0.0988, 7: 0.0988},
    10: {0: 0.1986, 1: 0.2003, 2: 0.1002, 3: 0.1002, 4: 0.1002,
         5: 0.1002, 6: 0.1002, 7: 0.1002},
    11: {0: 0.1887, 1: 0.2028, 2: 0.1014, 3: 0.1014, 4: 0.1014,
         5: 0.1014, 6: 0.1014, 7: 0.1014},
    12: {0: 0.1797, 1: 0.2051, 2: 0.1025, 3: 0.1025, 4: 0.1025,
         5: 0.1025, 6: 0.1025, 7: 0.1025},

    13: {0: 0.1716, 1: 0.1506, 2: 0.0753, 3: 0.0753, 4: 0.0753,
         5: 0.0753, 6: 0.0753, 7: 0.0753, 8: 0.0753, 9: 0.0753, 10: 0.0753},
    14: {0: 0.1641, 1: 0.1520, 2: 0.0760, 3: 0.0760, 4: 0.0760,
         5: 0.0760, 6: 0.0760, 7: 0.0760, 8: 0.0760, 9: 0.0760, 10: 0.0760},
    15: {0: 0.1573, 1: 0.1532, 2: 0.0766, 3: 0.0766, 4: 0.0766,
         5: 0.0766, 6: 0.0766, 7: 0.0766, 8: 0.0766, 9: 0.0766, 10: 0.0766},
    16: {0: 0.1510, 1: 0.1544, 2: 0.0772, 3: 0.0772, 4: 0.0772,
         5: 0.0772, 6: 0.0772, 7: 0.0772, 8: 0.0772, 9: 0.0772, 10: 0.0772},
}

# Spell costs (with Weaken)
SpellCosts = {
    "Wind":   {"Air": 1, "Mind": 1},
    "Water":  {"Air": 1, "Mind": 1, "Water": 1},
    "Earth":  {"Air": 1, "Mind": 1, "Earth": 2},
    "Fire":   {"Air": 2, "Mind": 1, "Fire": 3},
    "Weaken": {"Earth": 2, "Water": 3, "Body": 1},
}

# unlock levels for each spell (base magic level)
SPELL_UNLOCK_LEVEL = {
    "Wind":   1,
    "Water":  5,
    "Earth":  9,
    "Weaken": 11,
    "Fire":   13,
}

# Cum XP for both RC and Magic
xp_for_level = {
    1: 0,    2: 83,    3: 174,   4: 276,   5: 388,
    6: 512,  7: 650,   8: 801,   9: 969,   10: 1154,
    11: 1358, 12: 1584, 13: 1833, 14: 2107, 15: 2411,
    16: 2746, 17: 3115, 18: 3520, 19: 3973, 20: 4470,
    21: 5018, 22: 5624, 23: 6291, 24: 7028, 25: 7842,
    26: 8740, 27: 9730, 28: 10824, 29: 12031, 30: 13363,
    31: 14833, 32: 16456, 33: 18247, 34: 20224, 35: 22406,
    36: 24815, 37: 27473, 38: 30403, 39: 33648, 40: 37224,
    41: 41171, 42: 45529, 43: 50343, 44: 55649, 45: 61512,
    46: 67983, 47: 75127, 48: 83021, 49: 91721, 50: 101333,
}

# Rune yields per essence for RC (with Body from your table)
rune_yields = {
    "1-9":    {"Body": 1.50, "Fire": 3.00,  "Earth": 6.00,  "Water": 12.00, "Mind": 25.00, "Air": 50.00, "XP_Ess": 9.39},
    "10-19":  {"Body": 6.00, "Fire": 12.00, "Earth": 24.00, "Water": 21.00, "Mind": 18.00, "Air": 15.00, "XP_Ess": 10.52},
    "20-29":  {"Body": 16.00,"Fire": 15.00, "Earth": 14.00, "Water": 13.50, "Mind": 13.00, "Air": 12.00, "XP_Ess": 11.34},
    "30-39":  {"Body": 13.00,"Fire": 12.00, "Earth": 11.00, "Water": 9.00,  "Mind": 8.00,  "Air": 7.00,  "XP_Ess": 12.27},
    "40-49":  {"Body": 10.00,"Fire": 8.00,  "Earth": 7.50,  "Water": 7.00,  "Mind": 6.50,  "Air": 6.00,  "XP_Ess": 12.90},
    "50-59":  {"Body": 0.00, "Fire": 7.00,  "Earth": 6.50,  "Water": 6.00,  "Mind": 5.50,  "Air": 5.00,  "XP_Ess": 13.47},
}

# Zamorak Crafter drops
drop_table = [
    {"item": "Air rune",     "qty": 5,  "prob": 1/7},
    {"item": "Water rune",   "qty": 5,  "prob": 1/7},
    {"item": "Fire rune",    "qty": 2,  "prob": 1/7},
    {"item": "Body rune",    "qty": 3,  "prob": 1/14},
    {"item": "Chaos rune",   "qty": 4,  "prob": 1/14},
    {"item": "Earth rune",   "qty": 6,  "prob": 1/14},
    {"item": "Blood rune",   "qty": 4,  "prob": 1/35},
    {"item": "Death rune",   "qty": 3,  "prob": 1/35},
    {"item": "Soul rune",    "qty": 3,  "prob": 1/35},
    {"item": "Nature rune",  "qty": 2,  "prob": 1/35},
    {"item": "Law rune",     "qty": 3,  "prob": 1/70},
    {"item": "Coins",        "qty": 50, "prob": 1/7},
    {"item": "Pure essence", "qty": 3,  "prob": 1/14},
    {"item": "Pure essence", "qty": 10, "prob": 1/70},
]

# ----------------------------------------------------------------------
# Helper functions
# ----------------------------------------------------------------------

def lore_effective_level(lvl: int) -> int:
    return math.floor(1.1 * lvl)

def might_effective_level(lvl: int) -> int:
    return math.floor(1.15 * lvl)

def apply_prayer(level: int, prayer: str | None) -> int:
    if prayer == "lore":
        return lore_effective_level(level)
    elif prayer == "might":
        return might_effective_level(level)
    return level

def rc_band(level: int) -> dict:
    if 1 <= level <= 9:
        return rune_yields["1-9"]
    elif 10 <= level <= 19:
        return rune_yields["10-19"]
    elif 20 <= level <= 29:
        return rune_yields["20-29"]
    elif 30 <= level <= 39:
        return rune_yields["30-39"]
    elif 40 <= level <= 49:
        return rune_yields["40-49"]
    else:
        return rune_yields["50-59"]

def sample_crafter_drop() -> dict:
    r = random.random()
    cum = 0.0
    for d in drop_table:
        cum += d["prob"]
        if r < cum:
            return d
    return drop_table[-1]

def sample_damage(spell: str, eff_level: int) -> int:
    """
    Sample damage roll for a given spell and effective level.
    - Earth uses EarthLvl overrides where available.
    - Other spells use generic Lvl table.
    """
    eff_level = max(1, min(18, eff_level))

    if spell == "Earth" and eff_level in EarthLvl:
        dist = EarthLvl[eff_level]
    else:
        dist = Lvl[eff_level]

    r = random.random()
    cum = 0.0
    for dmg, p in dist.items():
        cum += p
        if r < cum:
            return dmg
    return max(dist.keys())

def best_available_spell(level: int, rune_inv: dict) -> str | None:
    """
    Strongest spell unlocked by base level AND affordable.
    Priority (per your latest request):
        Weaken > Earth > Fire > Water > Wind
    """
    priority = ["Weaken", "Earth", "Fire", "Water", "Wind"]
    for spell in priority:
        if level < SPELL_UNLOCK_LEVEL[spell]:
            continue
        costs = SpellCosts[spell]
        if all(rune_inv.get(r, 0.0) >= c for r, c in costs.items()):
            return spell
    return None

def level_from_xp(xp: float) -> int:
    level = 1
    for lvl in sorted(xp_for_level.keys()):
        if xp >= xp_for_level[lvl]:
            level = lvl
        else:
            break
    return level

def craft_essence_batch(ess_batch: int,
                        rc_level: int,
                        rc_xp: float,
                        rune_inv: dict,
                        crafted_runes: dict) -> tuple[int, float]:
    band = rc_band(rc_level)
    rc_xp += ess_batch * band["XP_Ess"]

    while rc_level < max(xp_for_level.keys()) and rc_xp >= xp_for_level[rc_level + 1]:
        rc_level += 1

    for r in ["Air", "Mind", "Water", "Earth", "Fire", "Body"]:
        gained = ess_batch * (band[r] / 100.0)
        rune_inv[r] += gained
        crafted_runes[r] += gained

    return rc_level, rc_xp

def cast_until_stuck(magic_level: int,
                     magic_xp: float,
                     rune_inv: dict,
                     prayer: str | None,
                     used_runes: dict,
                     spell_counts: dict,
                     max_level: int = 15) -> tuple[int, float]:
    """
    Spend runes on Magic until no spell is affordable or we hit max_level.
    Tracks spell_counts per spell.
    """
    while magic_level < max_level:
        spell = best_available_spell(magic_level, rune_inv)
        if spell is None:
            break

        # Pay cost
        costs = SpellCosts[spell]
        for r, c in costs.items():
            if rune_inv.get(r, 0.0) < c:
                return magic_level, magic_xp
        for r, c in costs.items():
            rune_inv[r] -= c
            used_runes[r] += c

        # Count cast
        spell_counts[spell] += 1

        # XP gain
        if spell == "Weaken":
            xp_gain = WEAKEN_XP
        else:
            eff_level = apply_prayer(magic_level, prayer)
            dmg = sample_damage(spell, eff_level)
            xp_gain = SPELL_XP_TABLE[spell][dmg]

        magic_xp += xp_gain

        while magic_level < max_level and magic_xp >= xp_for_level[magic_level + 1]:
            magic_level += 1

    return magic_level, magic_xp

# ----------------------------------------------------------------------
# Main simulation
# ----------------------------------------------------------------------

def simulate_to_15(prayer: str | None = None,
                   kills_per_cycle: int = 50,
                   ess_per_trip: int = 10,
                   kill_cap: int = 100_000) -> dict:

    magic_level = 1
    magic_xp = 0.0

    rc_level = 1
    rc_xp = 0.0

    rune_inv = {"Air": 0.0, "Mind": 0.0, "Water": 0.0,
                "Earth": 0.0, "Fire": 0.0, "Body": 0.0}
    essence = 0
    kills = 0

    dropped_runes = {"Air": 0.0, "Mind": 0.0, "Water": 0.0,
                     "Earth": 0.0, "Fire": 0.0, "Body": 0.0}
    crafted_runes = {"Air": 0.0, "Mind": 0.0, "Water": 0.0,
                     "Earth": 0.0, "Fire": 0.0, "Body": 0.0}
    used_runes    = {"Air": 0.0, "Mind": 0.0, "Water": 0.0,
                     "Earth": 0.0, "Fire": 0.0, "Body": 0.0}

    spell_counts  = {"Wind": 0, "Water": 0, "Earth": 0, "Fire": 0, "Weaken": 0}

    # NEW: track total pure essence dropped from crafters
    ess_dropped = 0

    while magic_level < 17 and kills < kill_cap:
        # 1) Kill crafters
        for _ in range(kills_per_cycle):
            kills += 1
            drop = sample_crafter_drop()
            item, qty = drop["item"], drop["qty"]

            if item == "Air rune":
                rune_inv["Air"] += qty
                dropped_runes["Air"] += qty
            elif item == "Water rune":
                rune_inv["Water"] += qty
                dropped_runes["Water"] += qty
            elif item == "Fire rune":
                rune_inv["Fire"] += qty
                dropped_runes["Fire"] += qty
            elif item == "Earth rune":
                rune_inv["Earth"] += qty
                dropped_runes["Earth"] += qty
            elif item == "Body rune":
                rune_inv["Body"] += qty
                dropped_runes["Body"] += qty
            elif item.startswith("Pure essence"):
                essence += qty
                ess_dropped += qty   # ← NEW

            if kills >= kill_cap:
                break

        # 2) RC trips
        while essence >= ess_per_trip:
            essence -= ess_per_trip
            rc_level, rc_xp = craft_essence_batch(
                ess_per_trip, rc_level, rc_xp, rune_inv, crafted_runes
            )

        # 3) Spend all possible runes
        magic_level, magic_xp = cast_until_stuck(
            magic_level, magic_xp, rune_inv, prayer, used_runes,
            spell_counts, max_level=17
        )

        if magic_level >= 17:
            break

    # floor runes for reporting
    for r in rune_inv:
        rune_inv[r]      = math.floor(rune_inv[r])
        dropped_runes[r] = math.floor(dropped_runes[r])
        crafted_runes[r] = math.floor(crafted_runes[r])
        used_runes[r]    = math.floor(used_runes[r])

    return {
        "kills": kills,
        "magic_level": magic_level,
        "magic_xp": magic_xp,
        "rc_level": rc_level,
        "rc_xp": rc_xp,
        "inventory_runes": rune_inv,
        "essence_left": essence,
        "dropped_runes": dropped_runes,
        "crafted_runes": crafted_runes,
        "used_runes": used_runes,
        "spell_counts": spell_counts,
        "ess_dropped": ess_dropped,   # ← NEW
    }


# ----------------------------------------------------------------------
# Multi-sim runner with histogram and spell stats
# ----------------------------------------------------------------------

def percentile(data, p):
    """Simple percentile helper, p in [0,100]."""
    if not data:
        return None
    s = sorted(data)
    k = (len(s) - 1) * p / 100.0
    f = math.floor(k)
    c = math.ceil(k)
    if f == c:
        return s[int(k)]
    return s[f] + (s[c] - s[f]) * (k - f)

def run_many_sims(n: int = 1000, prayer: str | None = None, show_hist: bool = True):
    kills_list = []
    mind_used_list = []
    total_spell_counts = {"Wind": 0, "Water": 0, "Earth": 0, "Fire": 0, "Weaken": 0}

    body_gained_list = []     # NEW
    ess_dropped_list = []     # NEW

    for _ in range(n):
        res = simulate_to_15(prayer=prayer)
        kills_list.append(res["kills"])
        mind_used_list.append(res["used_runes"]["Mind"])

        # spells
        for sp, cnt in res["spell_counts"].items():
            total_spell_counts[sp] += cnt

        # body runes gained = dropped + crafted
        body_gained = (
            res["dropped_runes"]["Body"] +
            res["crafted_runes"]["Body"]
        )
        body_gained_list.append(body_gained)

        # essence dropped
        ess_dropped_list.append(res["ess_dropped"])

    mean_kills = sum(kills_list) / n
    mean_mind  = sum(mind_used_list) / n
    mean_body  = sum(body_gained_list) / n
    mean_ess   = sum(ess_dropped_list) / n

    mean_spells = {sp: total_spell_counts[sp] / n for sp in total_spell_counts}

    print(f"Prayer={prayer}, sims={n}")
    print(f"  Avg kills to 17: {mean_kills:.2f}")
    print(f"  Avg Mind runes used: {mean_mind:.2f}")
    print(f"  Avg Body runes gained (drop+RC): {mean_body:.2f}")   # NEW
    print(f"  Avg pure essence dropped: {mean_ess:.2f}")          # NEW
    print("  Avg casts per spell:")
    for sp in ["Weaken", "Earth", "Fire", "Water", "Wind"]:
        print(f"    {sp:6s}: {mean_spells[sp]:.2f}")

    print("  Kill distribution:")
    print(f"    5th percentile : {percentile(kills_list, 5):.0f}")
    print(f"    25th percentile: {percentile(kills_list, 25):.0f}")
    print(f"    Median         : {percentile(kills_list, 50):.0f}")
    print(f"    75th percentile: {percentile(kills_list, 75):.0f}")
    print(f"    95th percentile: {percentile(kills_list, 95):.0f}")

    if show_hist:
        plt.figure()
        plt.hist(kills_list, bins=100)
        plt.xlabel("Kills to reach Magic 17")
        plt.ylabel("Frequency")
        plt.title(f"Distribution of kills to 17 (prayer={prayer}, n={n})")
        plt.tight_layout()
        plt.show()


# Example usage:
run_many_sims(1000000, prayer=None)
# run_many_sims(2000, prayer="lore")
# run_many_sims(2000, prayer="might")
````
